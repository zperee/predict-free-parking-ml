# Parkhäuser
Die Regelmässigkeit der Auslastung ist extrem unterschiedlich für die einzelnen Parkhäuser. Die einen scheinen sehr regelmässig zu sein, es gibt aber auch welche die erscheinen ganz und gar nicht regelmässig. Unten aufgelistet sind einige Beispiele. Das letzte Parkhaus ist auch jenes welches das Projektteam dachte, es könnten gute Resultate erziehlt werden, da es eines der regelmässigsten war. 

```python
# Daten visualisieren für alle Parkhäuser
for parkingid in df['parking_id'].unique():
    parking = by_hour[by_hour['parking_id'] == parkingid]
    print_timeline(parking, parkingid)
```
